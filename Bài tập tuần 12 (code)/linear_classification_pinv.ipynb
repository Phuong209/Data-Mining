```python
# train a linear regrssion model using Pseudo inverse of A

import numpy as np

A = [[40, 30, 1],
     [30, 80, 1],
     [20, 70, 1],
     [70, 40, 1],
     [60, 20, 1]]

Z = [[1], [-1], [-1], [1], [1]]
```

```python
At = np.transpose(A)

print(At)
```

```python
AtA = np.dot(At,A)
AtZ = np.dot(At,Z)

print(AtA)
print(AtZ)
```

```python
inv_AtA = np.linalg.inv(AtA)
X = np.dot(inv_AtA,AtZ)
```

```python
print(X)
```
