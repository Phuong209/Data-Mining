```python
training_data = [[170, 65],
                 [155, 50],
                 [150, 45],
                 [175, 70],
                 [165, 55]]
```

```python
print(training_data)
```

```python
import sympy as sym

a = sym.Symbol('a')
b = sym.Symbol('b')

Residual = 0

for i in range(len(training_data)):
    
    Residual += (training_data[i][0]*a + b - training_data[i][1]) **2
```

```python
Residual
```

```python
R_Diff_a = sym.diff(Residual,a)
R_Diff_a
```

```python
R_Diff_b = sym.diff(Residual,b)
R_Diff_b
```
