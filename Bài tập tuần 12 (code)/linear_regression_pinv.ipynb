```python
# train a linear regrssion model using Pseudo inverse of A

import numpy as np

A = [[170, 1],
     [155, 1],
     [150, 1],
     [175, 1],
     [165, 1]]

Y = [[65], [50], [45], [70], [55]]
```

```python
At = np.transpose(A)

print(At)
```

```python
AtA = np.dot(At,A)
AtY = np.dot(At,Y)

print(AtA)
print(AtY)
```

```python
inv_AtA = np.linalg.inv(AtA)
X = np.dot(inv_AtA,AtY)
```
