#### 2. Các thao tác trên dãy số a
(a) Bình phương các phần tử trong dãy số <br/>
(b) Tìm số lượng các phần tử của dãy số <br/>
(c) In ra giá trị lớn nhất, giá trị nhỏ nhất và giá trị trung bình <br/>
(d) Tính phương sai của dãy số $\sigma^2 = \frac{\sum{\left(a_i - \mu\right)^2}}{N - 1}$ với $\mu = \frac{\sum{a_i}}{N}$ và độ lệch chuẩn $var = \sqrt{\sigma^2}$

```python
import numpy as np

print("2. Các thao tác tính toán đơn giản với dãy số a")

# Giả sử a đã có từ trước (ví dụ a = np.arange(1, 11))
a = np.arange(1, 11)

# (a) Bình phương các phần tử trong dãy số
a2 = a**2

# (b) Tìm số lượng các phần tử của dãy số
na = len(a)

# (c) In ra giá trị lớn nhất, giá trị nhỏ nhất và giá trị trung bình
v_max = np.max(a)
v_min = np.min(a)
v_avg = np.mean(a)

# (d) Tính phương sai và độ lệch chuẩn
mu = np.mean(a)                        # trung bình
v_sigma2 = np.sum((a - mu)**2) / (na-1) # phương sai mẫu
v_var = np.sqrt(v_sigma2)              # độ lệch chuẩn

print("a. Binh phuong a: ", a2)
print("\nb. Do dai a: ", na)
print(f'\nc. Gia tri lon nhat: [{v_max}], gia tri nho nhat: [{v_min}], gia tri trung binh [{v_avg:.2f}]')
print(f'\nd. Phuong sai: [{v_sigma2:.2f}] va do lech chuan: [{v_var:.2f}]')
```

#### 3. Toán tử trên hai dãy số
(a) Tạo dãy số x có 10 phần tử từ 1 đến 10 (dùng hàm np.arange) <br/>
(b) Liệt kê và đếm số lượng phần tử chẵn và lẻ có trong dãy a và x <br/>
(c) Tính khoảng cách giữa các phần tử ở vị trí lẻ của mảng a và x với $s = \sqrt{\sum{\left(a_i - x_i\right)^2}}$ với $i$ lẻ<br/>
(d) Tìm khoảng cách nhỏ nhất giữa tập a và x. Ví dụ: min({1, 3, 5}, {2, 10, 8}} = 1

```python
print("3. Xử lý khác trên dãy số")

# 3a. Tạo dãy số x có 10 phần tử từ 1 đến 10
import numpy as np
x = np.arange(1, 11)

print(f'a. Day so x: {x}')
```

```python
""" CÁC BẠN LÀM BÀI Ở ĐÂY """
# 3b. Lọc các phần tử chẵn trong dãy a và x

# Tạo điều kiện lọc (True nếu là số chẵn, False nếu là lẻ)
a_la_chan = (a % 2 == 0)
x_la_chan = (x % 2 == 0)

# Lấy ra các phần tử chẵn
a_chan = a[a_la_chan]
x_chan = x[x_la_chan]

# Đếm số lượng phần tử chẵn
n_a_chan = len(a_chan)
n_x_chan = len(x_chan)

print("b. Cac phan tu chan trong day a, b: ")
print(f"+ Day a: {a}")
print(f"  - Loc chan: {a_la_chan}")
print(f"  - Day chua phan tu chan trong a: {a_chan}, so luong {n_a_chan} phan tu.")
print(f"+ Day x: {x}")
print(f"  - Loc chan: {x_la_chan}")
print(f"  - Day chua phan tu chan trong a: {x_chan}, so luong {n_x_chan} phan tu.")
```

```python
""" CÁC BẠN LÀM BÀI Ở ĐÂY """
# 3c. Tính hiệu và khoảng cách giữa các phần tử ở vị trí lẻ của mảng a và x

# Lấy chỉ số lẻ
v_le = np.arange(1, len(a), 2)

# Lấy các phần tử ở vị trí lẻ
a_le = a[v_le]
x_le = x[v_le]

# Tính hiệu
hieu_le = a_le - x_le

# Tính khoảng cách
kc_le = np.sqrt(np.sum((a_le - x_le) ** 2))

print("c. Tinh hieu cac phan tu o vi tri le cua mang a va x: ")
print(f"+ Vi tri le trong a va x: {v_le}")
print(f"+ Day so a: {a} va cac so o vi tri le: {a_le}")
print(f"+ Day so x: {x} va cac so o vi tri le: {x_le}")
print(f"+ Hieu cua a va x o vi tri le: {hieu_le}")
print(f"+ Khoang cach cac phan tu cua a va x o vi tri le: {kc_le: .2f}")
```

```python
print("d. Tìm khoảng cách nhỏ nhất giữa tập a và x. Ví dụ: min({1, 3, 5}, {2, 10, 8}} = 1")
"""
Gợi ý: Áp dụng toán tử broadcasting
[[1],                    [[1, 1, 1],        [[2, 10, 8],    =   [[-1, -9, -7],
 [3],   - [2, 10, 8] =    [3, 3, 3],   -     [2, 10, 8],         [ 1, -7, -5],  --> np.min(np.abs(...)) = 1
 [5]]                     [5, 5, 5]]         [2, 10, 8]]         [ 3, -5, -3]]
"""
""" CÁC BẠN LÀM BÀI Ở ĐÂY """
# Tăng số chiều dùng tại vị trí chiều cuối dùng np.newaxis như sau: a[:, np.newaxis]
kc = a[:, np.newaxis] - x     # ma trận khoảng cách
kc_abs = np.abs(kc)           # trị tuyệt đối
v_min = np.min(kc_abs)        # lấy khoảng cách nhỏ nhất

print(f"+ Day a: {a}")
print(f"+ Day x: {x}")
print(f"+ Khoang cach nho nhat: {v_min}")
```

```python
""" CÁC BẠN LÀM BÀI Ở ĐÂY """
# Sử dụm hàm np.random.rand() sinh số ngẫu nhiên trong [0,1) --> [0,1) * (b - a) + a --> [a, b)
# Sử dụng hàm sau để tính hàm xác suất chuẩn N(m, s): np.sqrt, np.exp
mu, sigma = 1.0, 1.0
N = 100
x_min, x_max = mu - 5 * sigma, mu + 5 * sigma
x = np.random.rand(N) * (x_max - x_min) + x_min
y = 1 / (sigma * np.sqrt(2 * np.pi)) * np.exp(-((x - mu) ** 2) / (2 * sigma ** 2))


print("b. Sinh cặp (x, y) với y làm hàm phân phối chuẩn")
print(f"+ x: {x_min} <= min [{np.min(x): .2f}] <= max [{np.max(x): .2f}] <= {x_max}")
print(f"+ Ve ham so y(x)")
plt.figure(figsize=(6,6))      # kich thuoc ban ve
plt.scatter(x, y)              # ve cac diem (xi, yi)
plt.savefig("4b.png")          #  luu do thi
plt.show()                     #  hien thi do thi
```

```python
""" CÁC BẠN LÀM BÀI Ở ĐÂY """
# Sử dụng hàm np.random.randn() sinh phân phối chuẩn N(0,1) --> N(m,s^2) = m + N(0,1) * s
# Sử dụng hàm np.histogram với density=True để thống kê số lần xuất hiện với số bins = 300 (do dãy số thực)
mu, sigma = 1.0, 1.0
N = 100000
x = mu + np.random.randn(N) * sigma
cnt, val = np.histogram(x, bins=300, density=True)
val = (val[:-1] + val[1:]) / 2   # lấy tâm mỗi khoảng để vẽ


print( "c. Sinh day so theo phan phoi chuan")
print(f"+ x: len = {len(x)}, min [{np.min(x)}], max [{np.max(x)}]")
print(f"+ Mat so xac suat cua day so x")
plt.figure(figsize=(6,6))      # kich thuoc ban ve
plt.xlim(np.min(val), np.max(val))
plt.scatter(val, cnt)
plt.show()
```

#### 4. Giải hệ phương trình tuyến tính
Giải hệ phương trình tuyến tính sau:
$$
\left\{\begin{matrix}
        4x_1  + 3x_2 - 5x_3 = 2 \\
        -2x_1 - 4x_2 + 5x_3 = 5 \\
        8x_1 + 8x_2 = -3 \\
\end{matrix}\right.
$$

```python
import numpy as np

A = np.array([
    [4, 3, -5],
    [-2, -4, 5],
    [8, 8, 0]
], dtype=float)

y = np.array([2, 5, -3], dtype=float)

x = np.linalg.solve(A, y)

yy = A @ x
diff = np.linalg.norm(yy - y, 2)

print(f"+ Ma tran A: \n{A}")
print(f"+ Vector y: {y}")
print(f"+ Nghiem x: {x}")
print(f"+ Kiem tra: yy = Ax = {yy} ==> ||yy - y||_2 = {diff:.2f}")
```

### Bài 3. Tính gần đúng số pi dùng phương pháp Monte Carlo

Vẽ một đường tròn $C$ có bán kính $r$ nội tiếp một hình vuông $C$, và đặt ngẫu nhiên $N$ chấm lên hình vuông. Tỉ lệ các chấm nằm trong hình tròn trên tổng số chấm xấp xỉ bằng diện tích của hình tròn chia cho hình vuông như sau:
$$
\frac{N_C}{N} \approx  \frac{\mathbb{S}_C}{\mathbb{S}_N}
$$
nơi mà $N_C$ là các chấm trong hình tròn, $N$ là tổng số chấm, $\mathbb{S}_C$ và $\mathbb{S}_N$ lần lượt là diện tích của hình tròn và hình vuông.
$$ \mathbb{S}_C = \pi * r^2 $$
$$ \mathbb{S}_N = (2*r)^2 = 4r^2 $$
Do đó, công thức trên trở thành:
$$
\frac{N_C}{N} \approx \frac{\pi * r^2}{4r^2} = \frac{\pi}{4}
$$
Suy ra, ta có giá trị số $\pi$ được tính như sau:
$$
\pi \approx \frac{4N_C}{N}
$$

Các bạn hãy lập trình tính sấp xỉ số $\pi$ dùng phương pháp Monte Carlo. Cho biết các sai số với $N = 100$, $N = 10000$, và $N = 1000000$. Vẽ đồ thị minh họa cho chương trình.

**Hướng dẫn**: Viết chương trình phát sinh $N$ điểm $(x_i, y_i)$ với $x_i, y_i \in [-1, -1]$. $(x_i, y_i)$ thuộc hình tròn tâm $(0, 0)$ bán kính 1 nội tiếp hình vuông có tâm tại $(0, 0)$ độ dài cạnh 2 khi và chỉ khi $x_i^2 + y_i^2 <= 1$.

```python
import matplotlib.pyplot as plt # khai báo thư viện vẽ pyplot
import math

circle = plt.Circle((0, 0), 1, color='g') # tạo đối tượng vòng tròn

fig, ax = plt.subplots(figsize=(6, 6)) # lấy figure và vùng vẽ

plt.plot(0, 0, 'o', color=(0.9, 0.9, 1.0), alpha=0.8) # vẽ điểm tại tâm (0,0)
ax.add_patch(circle) # thêm vòng tròn vào vùng vẽ

x = 0.75; y = math.sqrt(1 - (x ** 2)) # (x = 0.75, y = căn bậc 2 của 1 - x^2)
plt.arrow(0, 0, x, y)

plt.xlim(-1, 1) # giới hạn trục x
plt.ylim(-1, 1) # giới hạn trục y

plt.savefig('pi.png') # lưu đồ thị xuống tập tin
plt.show()            # hiển thị đồ thị
```

```python
import numpy as np
import matplotlib.pyplot as plt

def calc_pi_monte_carlo(n = 100):
    # sinh n điểm ngẫu nhiên trong hình vuông [-1,1] x [-1,1]
    x = np.random.rand(n) * 2 - 1   # từ [-1,1]
    y = np.random.rand(n) * 2 - 1   # từ [-1,1]

    # đếm số điểm nằm trong đường tròn bán kính 1 (x^2 + y^2 <= 1)
    inside = (x**2 + y**2) <= 1
    nc = np.sum(inside)

    # công thức ước lượng pi
    pi_est = 4 * nc / n
    return pi_est

# chạy thử với các N khác nhau
N_list = [100, 10000, 1000000]
pi_values = []
errors = []

for N in N_list:
    pi_est = calc_pi_monte_carlo(N)
    pi_values.append(pi_est)
    errors.append(abs(np.pi - pi_est))

    print(f"N = {N}, pi ~ {pi_est}, sai số = {abs(np.pi - pi_est)}")

# Vẽ biểu đồ sai số
plt.figure(figsize=(7,5))
plt.plot(N_list, errors, 'o-', label="Sai số |pi - pi_est|")
plt.xscale("log")
plt.xlabel("Số điểm N (log scale)")
plt.ylabel("Sai số")
plt.title("Sai số khi xấp xỉ pi bằng Monte Carlo")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import math

print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=100): ", calc_pi_monte_carlo(n = 100) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=10000): ", calc_pi_monte_carlo(n = 10000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
print("epsilon(n=1000000): ", calc_pi_monte_carlo(n = 1000000) - math.pi)
```
